```python
import pickle
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

# Features and their encoders
encoders = {
    'Brand': LabelEncoder(),
    'Model': LabelEncoder(),
    'UsedOrNew': LabelEncoder(),
    'Transmission': LabelEncoder(),
    'DriveType': LabelEncoder(),
    'FuelType': LabelEncoder(),
    'BodyType': LabelEncoder()
}
en=OneHotEncoder()
en.fit(df[encoders.keys()])
transformed_data = {}

for feature, encoder in encoders.items():
    
    # transformed_data[feature] = encoder.fit_transform(df[feature])
    df[feature]= encoder.fit_transform(df[feature])
    x_train[feature]=encoder.transform(x_train[feature])
    x_test[feature]=encoder.transform(x_test[feature])
 
    with open(f'{feature}_encoder.pkl', 'wb') as f:
        pickle.dump(encoder, f)

print("All encoders have been fit, transformed, and pickled successfully!")
```
